```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pickle
```

```python
res = pickle.load(open('./results_week6/results-cur.p', 'rb'))
```

```python
x = np.linspace(500, 100000, 20).astype(np.int)
ghat_c = []
ghat_uc = []

fr_c = []
fr_uc = []

acc_c = []
acc_uc = []
for r in res:
    fr_c.append(r['failure_rate'])
    fr_uc.append(r['uc_failure_rate'])
    
    acc_c.append(r['accuracy'])
    acc_uc.append(r['uc_accuracy'])
plt.plot(x, fr_c, label='Constrained failure rate')
plt.plot(x, fr_uc, label='Unconstrained failure rate')
plt.xlabel('Number of data points')
plt.ylabel('failure rate')
plt.legend()
plt.title('Plot for number of data points vs failure rate')
plt.show()
plt.plot(x, acc_c, label='Constrained accuracy')
plt.plot(x, acc_uc, label='Unconstrained accuracy')
plt.xlabel('Number of data points')
plt.ylabel('Accuracy')
plt.legend()
plt.title('Plot for number of data points vs accuracy')
plt.show()
```
